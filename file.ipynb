```python
import csv
data =[
    [ "Name", "Age", "City"],
    ["Pratima", "21", "Kathmandu"],
    ["Pranusha", "12","Kathmandu"]
]
with open('example.csv','w',newline='') as file:
    writer=csv.writer(file)
    writer.writerows(data)
    print("done")
```

```python
import csv
with open('example.csv','r') as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)
```

```python
import json
data = {
    "name": "prince",
    "age": "21",
    "city": "Nepal"
}
with open ('example.json','w') as file:
    json.dump(data,file,indent=4)
    print("done")
```

```python
import json
with open('example.json','r') as f:
    data=json.load(f)
    print(data)
```
